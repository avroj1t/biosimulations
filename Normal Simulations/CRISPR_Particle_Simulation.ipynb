# CRISPR PARTICLE SIMULATION

The code defines a simple class to simulate and visualize the movement of a CRISPR-Cas9 delivery particle over time. We have taken the sine function to simulate the particle behaviour

```python
import numpy as np
import matplotlib.pyplot as plt

class CRISPRParticleSimulation:
    def __init__(self, total_time=10, num_time_points=100):
        self.total_time = total_time
        self.num_time_points = num_time_points
        self.time_points = np.linspace(0, total_time, num_time_points)

    def simulate_particle_behavior(self):
        # Example simulation code for particle position (modify with actual simulation logic)
        particle_position = np.sin(self.time_points)  # Replace this with actual simulation logic
        return particle_position

    def visualize_simulation(self, particle_position):
        # Visualize the simulated particle behavior
        plt.figure(figsize=(8, 6))
        plt.plot(self.time_points, particle_position, color='b', label='Particle Position')
        plt.xlabel('Time')
        plt.ylabel('Position')
        plt.title('CRISPR-Cas9 Particle Behavior During Delivery')
        plt.legend()
        plt.show()

# Example usage
if __name__ == "__main__":
    particle_simulation = CRISPRParticleSimulation()
    simulated_position = particle_simulation.simulate_particle_behavior()
    particle_simulation.visualize_simulation(simulated_position)
```

# CRISPR CAS9 Particle Movement
For simulations involving CRISPR-Cas9 particles or similar, where particles are small and suspended in a medium, Brownian motion provides a good approximation of how these particles might behave over time.


```python
class CRISPRParticleSimulation:
    def __init__(self, total_time=10, num_time_points=100):
        self.total_time = total_time
        self.num_time_points = num_time_points
        self.time_points = np.linspace(0, total_time, num_time_points)

    def simulate_particle_behavior(self):
        displacements = np.random.randn(self.num_time_points)  # Random steps at each time point
        particle_position = np.cumsum(displacements)  # Cumulative sum to get the position at each point
        return particle_position

    def visualize_simulation(self, particle_position):
        # Visualize the simulated particle behavior (Brownian motion)
        plt.figure(figsize=(8, 6))
        plt.plot(self.time_points, particle_position, color='b', label='Particle Position')
        plt.xlabel('Time')
        plt.ylabel('Position')
        plt.title('Brownian Motion of CRISPR-Cas9 Particle')
        plt.legend()
        plt.show()

# Example usage
if __name__ == "__main__":
    particle_simulation = CRISPRParticleSimulation()
    simulated_position = particle_simulation.simulate_particle_behavior()
    particle_simulation.visualize_simulation(simulated_position)
```
